# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
# Verificando a tabela para ser realizado os calculos necessários

dados.head()
```

```python
# Calculando a média dos valores presentes da coluna 'Valor'

dados['Valor'].mean()
```

```python
# Agrupamento e Cálculo da Média por Tipo de Imóvel
# O parâmetro numeric_only=True garante que apenas colunas numéricas sejam incluídas no cálculo.

dados.groupby('Tipo').mean(numeric_only=True)
```

```python
# Agrupamento e Cálculo da Média por Tipo e Valor medio do imovel
# Exibindo so o tipo a a media dos valores

dados.groupby('Tipo')['Valor'].mean()
```

```python
# Agrupar os dados pelo tipo de imóvel, calcular a média dos valores e ordenar
# O resultado é ordenado de forma ascendente pela média dos valores usando sort_values('Valor').

dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Agrupamento e Cálculo da Média: O comando dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') agrupa os dados do DataFrame dados pela coluna Tipo, calcula a média dos valores da coluna Valor para cada tipo de imóvel e ordena os resultados de forma ascendente. O resultado é armazenado na variável df_preco_tipo.
# O método plot(kind='barh', figsize=(14, 10), color='purple') é utilizado para criar um gráfico de barras horizontal a partir do DataFrame df_preco_tipo. Os parâmetros especificam:

# kind='barh': Indica que o gráfico deve ser do tipo de barras horizontal.
# figsize=(14, 10): Define o tamanho da figura em polegadas (largura x altura).
# color='purple': Define a cor das barras como roxa.

df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='purple');
```

## Removendo os imóveis comerciais

```python
# Identificação de Tipos Únicos de Imóveis

dados.Tipo.unique()
```

```python
# Selecionando apenas os imoveis comerciais

imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
# O comando `dados.query('@imoveis_comerciais in Tipo')` é utilizado para filtrar o DataFrame `dados`, retornando apenas as linhas onde o tipo de imóvel corresponde a uma lista ou conjunto de tipos especificados na variável `imoveis_comerciais`.

dados.query('@imoveis_comerciais in Tipo')
```

```python
# O comando `dados.query('@imoveis_comerciais not in Tipo')` é utilizado para filtrar o DataFrame `dados`, retornando apenas as linhas onde o tipo de imóvel **não** corresponde a uma lista ou conjunto de tipos especificados na variável `imoveis_comerciais`.

dados.query('@imoveis_comerciais not in Tipo')
```

```python
# Cálculo da Média de Preços por Tipo e Visualização Gráfica

df_preco_tipo = df_imoveis.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='purple');
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
# Indentificando o tipo unico dos imoveis não comerciais

df_imoveis.Tipo.unique()
```

```python
# O comando `df_imoveis.Tipo.value_counts(normalize=True)` é utilizado para contar a frequência de cada tipo de imóvel no DataFrame `df_imoveis`. A contagem é normalizada, ou seja, os valores são convertidos em proporções em relação ao total, facilitando a análise relativa entre os diferentes tipos de imóveis.

df_imoveis.Tipo.value_counts(normalize=True)
```

```python
# O comando `df_imoveis.Tipo.value_counts(normalize=True).to_frame().sort_values(by='proportion', ascending=False)` é utilizado para contar a frequência de cada tipo de imóvel no DataFrame `df_imoveis`, normalizando esses valores para proporções e, em seguida, ordenando os resultados em ordem decrescente.

df_imoveis.Tipo.value_counts(normalize=True).to_frame().sort_values(by='proportion', ascending=False)
```

```python
# Gráfico feito do percentual dos tipos de imoveis não comerciais

df_percentual_tipo = df_imoveis.Tipo.value_counts(normalize=True).to_frame().sort_values(by='proportion', ascending=False)

df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color ='green',
                        xlabel = 'Tipos', ylabel = 'Percentual');
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
# Visualizando a coluna quartos

df_apartamento['Quartos']
```

```python
# Passando a seleção de quartos(quantidade = 1) para  variável

selecao_quartos = df_apartamento['Quartos'] ==1
df_apartamento[selecao_quartos]
```

```python
# Passando a seleção dos valores para a variável

selecao_valor = df_apartamento['Valor'] < 1200
df_apartamento[selecao_valor]
```

```python
# Passando a seleção das duas ultimas consultas para variável

selecao_final = (selecao_quartos & selecao_valor)
df_apartamento[selecao_final]
```

```python
# Salvando os dois filtros da seleção final

df_filtro_quarto_e_valor = df_apartamento[selecao_final]
```

```python
# Visualizando

df_filtro_quarto_e_valor
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
# Fazendo a segunda seleção e passando para uma variável, essa seleção é composta

selecao_2 = (df_apartamento['Quartos'] >= 2) & (df_apartamento['Valor'] < 3000) & (df_apartamento['Area'] > 70)
df_apartamento[selecao_2]
```

```python
# Salvando a seleção nessa variável

df_filtro_quarto_valor_e_area = df_apartamento[selecao_2]
```

```python
# Visualizando a seleção

df_filtro_quarto_valor_e_area
```

# Manipulando os dados

## Criando colunas numéricas

```python
# Visualizando os dados la do inicio

dados
```

```python
# Incluindo outra coluna com a soma do valor + condominio

dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
# incluindo outra coluna com a multiplicação do valor*12 + IPTU

dados['Valor_por_ano'] = dados['Valor'] *12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
# Criando colunas categoricas que aplicam descrição

dados['Descricao'] = dados['Tipo'] + ' localizada em ' + dados['Bairro']
dados.head()
```

```python
# Aplicando mais descrição às colunas

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
# Avaliando se os quartos possuem suite ou não e criando uma nova coluna para isso

dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
# Salvando os dados e fazendo o download posteriormente

dados.to_csv('dados_completos.csv', index=False, sep=';')
```
